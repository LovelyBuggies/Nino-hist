```python
import hist
import numpy as np
```

```python
h = hist.Hist(
    hist.axis.Regular(100, -3, 3, name="S",title="s [units]", underflow=False, overflow=False)
)

data = np.random.normal(size=1_000)
h.fill(data)
h.view()
```

```python
def pdf(x):
    return 1/np.sqrt(2*np.pi) * np.exp(-.5*x**2)
```
